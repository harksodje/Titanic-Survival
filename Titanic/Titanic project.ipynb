```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import PolynomialFeatures
#loading train file
df = pd.read_csv('train.csv')
df1 = pd.DataFrame(df)
df1.isnull().sum()
df1["Age"].fillna(df1["Age"].median(),inplace = True)
df1["Embarked"].fillna(method ='ffill',inplace =True)
df1.drop(['Name','Ticket','Cabin'], axis =1, inplace = True)
df1.head()
```

```python
lb= LabelEncoder()
df1['Embarked']=lb.fit_transform(df1['Embarked'],)
df1['Sex'] = lb.fit_transform(df1['Sex'])
df1.head(5)
```

```python
X = df1[['PassengerId', 'Pclass', 'Sex', 'SibSp','Parch','Embarked']].values
Y = df1['Survived'].values
ohe=OneHotEncoder(categorical_features=[2,5])
ohe.fit_transform(X)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X,Y)
```
